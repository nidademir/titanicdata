```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
```

1. Kazada ölenlerin yaş ortalamasını bulunuz

```python
olen_yas_ort = titanic.loc[titanic['survived'] == 0, 'age'].mean()
print("Kazada ölenlerin yaş ortalaması:", olen_yas_ort)
```

2. Kazada ölenlerin bilet fiyatlarının ortalamasını ve medyanını bulunuz

```python
olen_fiyat_ort = titanic.loc[titanic['survived'] == 0, 'fare'].mean()
olen_fiyat_medyan = titanic.loc[titanic['survived'] == 0, 'fare'].median()

print("Kazada ölenlerin bilet fiyatlarının ortalaması:", olen_fiyat_ort)
print("Kazada ölenlerin bilet fiyatlarının medyanı:", olen_fiyat_medyan)
```

3. Kazada ölen erkeklerin yaş ortalamasını bulunuz

```python
olen_erkek_yas_ort = titanic.loc[(titanic['survived'] == 0) & (titanic['sex'] == 'male'), 'age'].mean()

print("Kazada ölen erkeklerin yaş ortalaması:", olen_erkek_yas_ort)
```

4. Kazada ölen kadınların yaş ortalamasını bulunuz

```python
olen_kadin_yas_ort = titanic.loc[(titanic['survived'] == 0) & (titanic['sex'] == 'female'), 'age'].mean()

print("Kazada ölen kadınların yaş ortalaması:", olen_kadin_yas_ort)
```

5.Kazadan kurtulanların yaş ortalamasını bulunuz

```python
kurtulan_yas_ort = titanic.loc[titanic['survived'] == 1, 'age'].mean()

print("Kazadan kurtulanların yaş ortalaması:", kurtulan_yas_ort)
```

6.Kazadan kurtulanların bilet fiyatlarının ortalamasını bulunuz

```python
kurtulan_fiyat_ort = titanic.loc[titanic['survived'] == 1, 'fare'].mean()

print("Kazadan kurtulanların bilet fiyatlarının ortalaması:", kurtulan_fiyat_ort)
```

7. Kazadan kurtulan toplam kişi sayısını bulunuz…

```python
kurtulan_sayisi = titanic.loc[titanic['survived'] == 1, 'survived'].count()

print("Kazadan kurtulan toplam kişi sayısı:", kurtulan_sayisi)

# len ile

kurtulan_toplam = len(titanic[(titanic.survived==1)])

print("Kazadan kurtulan toplam kişi sayısı:", kurtulan_toplam)
```

8. 10 yaşından küçüklerin bilet fiyatlarının medyan değerini bulunuz

```python
on_yas_alt_fiyat_medyan = titanic.loc[titanic['age'] < 10, 'fare'].median()

print("10 yaşından küçüklerin bilet fiyatlarının medyan değeri:", on_yas_alt_fiyat_medyan)
```

9. 1.sınıf, 2.sınıf ve 3.sınıf bilet fiyatlarının ortalama ve medyanlarını karşılaştırınız. (Pclass değişkeni sınıfları barındırmaktadır.)

```python
sinif_bilgileri = titanic.groupby('pclass')['fare'].agg(['mean', 'median']).reset_index()

print("Bilet fiyatlarının sınıflara göre ortalama ve medyan değerleri:")
print(sinif_bilgileri)
```

10. Kazada ölen kadınların oranı ile erkeklerin oranını karşılaştırınız. (Örnek: erkekler için; ölen erkeklerin, erkek sayısına bölümü bu oranı vermektedir.

```python
toplam_erkek = titanic[titanic['sex'] == 'male'].shape[0]
toplam_kadin = titanic[titanic['sex'] == 'female'].shape[0]

olen_erkek = titanic[(titanic['survived'] == 0) & (titanic['sex'] == 'male')].shape[0]
olen_kadin = titanic[(titanic['survived'] == 0) & (titanic['sex'] == 'female')].shape[0]

oran_erkek = olen_erkek / toplam_erkek if toplam_erkek != 0 else None
oran_kadin = olen_kadin / toplam_kadin if toplam_kadin != 0 else None

print(f"Kadınların ölüm oranı: {oran_kadin:.2%}")
print(f"Erkeklerin ölüm oranı: {oran_erkek:.2%}")
```
